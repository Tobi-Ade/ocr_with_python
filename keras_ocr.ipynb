```python
import keras_ocr as kcr
import matplotlib.pyplot as plt
```

```python
pipeline = kcr.pipeline.Pipeline()
```

```python
images = [kcr.tools.read(img) for img in [
                                        "./images/Iconberg-Snapper-Frame-300x457.jpg",
                                        "./images/download.png",
                                        "./images/images.jfif",
                                    ]
        ]
```

```python
pred_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, pred_groups):
    kcr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

```python
# [text for (text, box) in predictions for prediction in pred_groups]
prediction = pred_groups[0]
for text, box in prediction:
    print(text)
```
