```python
import os
import keras_ocr as kcr
import matplotlib.pyplot as plt
```

```python
pipeline = kcr.pipeline.Pipeline()
```

```python
parent_dir = "/content/drive/MyDrive/Tobz/Data Science /ocr_with_python/images"

file_paths = os.listdir(parent_dir)

img_files = [os.path.join(parent_dir, file) for file in file_paths if file.endswith("jpg")]
```

```python
images = [kcr.tools.read(img) for img in img_files]
```

```python
pred_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, pred_groups):
    kcr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
